```python

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from datetime import date
from dateutil.relativedelta import relativedelta
import yfinance as yf
from datetime import datetime
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import LSTM, Dropout, Dense
from sklearn.model_selection import train_test_split
```

```python
start_date = date.today() - relativedelta(years=5)
end_date = date.today()

start_datetime = datetime(start_date.year, start_date.month, start_date.day)
end_datetime = datetime(end_date.year, end_date.month, end_date.day)

# Download data
historic_data = yf.download('^NSEI', start=start_datetime, end=end_datetime)

historic_data.index = historic_data.index.strftime('%Y-%m-%d')

historic_data

# 👉 Convert to image (show first 10 rows for readability)
#fig, ax = plt.subplots(figsize=(12, 5))
#ax.axis('tight')
#ax.axis('off')
#
#table = ax.table(
#    cellText=historic_data.head(10).values,
#    colLabels=historic_data.columns,
#    rowLabels=historic_data.head(10).index,
#    loc='center'
#)

#plt.show()
```

```python

sentiment_data = pd.read_csv('/content/drive/MyDrive/data sets for  ml/csv_files for creat dataset/final_sentiments.csv', usecols=["Date Created", "Mode"])
sentiment_data.columns = ['Date','Sentiment']
sentiment_data.set_index('Date',inplace=True)

sentiment_data
```

```python

historic_data.columns = historic_data.columns.get_level_values(0)

data = pd.merge(historic_data, sentiment_data, left_index=True, right_index=True)
data.dropna(inplace=True)

data
```
